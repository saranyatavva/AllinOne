```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")
```

```python
pic=cv2.imread("image.jpeg")
print(pic.shape)#1,401,175 PIXELS
plt.imshow(pic)
```

```python
ori_pic=cv2.cvtColor(pic,cv2.COLOR_BGR2RGB)
plt.imshow(ori_pic)
```

```python
color=("b","g","r")
for i,color in enumerate(color):
    histr=cv2.calcHist([pic],channels=[i],mask=None,histSize=[256],ranges=[0,256])
    plt.plot(histr,color=color)
    #plt.xlim([0,256])
plt.title("HISTOGRAM")
plt.xlabel("Frequency Values")
plt.ylabel("Pixel Intensity Values")
```

```python

b,g,r = cv2.split(ori_pic)
```

```python
equ_b = cv2.equalizeHist(b)
equ_g = cv2.equalizeHist(g)
equ_r = cv2.equalizeHist(r)
equ = cv2.merge((equ_b, equ_g, equ_r))
```

```python
plt.imshow(equ)
```
